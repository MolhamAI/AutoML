```python
def build_task5_statistics_A(metrics_dir, output_dir, baseline_model, metric_names):

    results = load_all_results(metrics_dir, data_dict_classification_only.keys(), AUTOML + BOOSTINGS)

    for dataset_name in data_dict_classification_only.keys():
        dataset_base = dataset_name.replace('.csv', '')
        df_all = []

        if baseline_model not in results[dataset_name]:
            continue

        baseline_df = results[dataset_name][baseline_model]

        for framework in AUTOML + BOOSTINGS:
            if framework == baseline_model:
                continue
            if framework not in results[dataset_name]:
                continue

            df_model = results[dataset_name][framework]

            row = {
                "dataset_name": dataset_base,
                "model": framework,
                "baseline": baseline_model
            }

            for metric in metric_names:
                m = df_model[metric].astype(float).values
                b = baseline_df[metric].astype(float).values

                mean_m = m.mean()
                std_m = m.std()
                ci_low, ci_high = bootstrap_confidence_interval(m)
                p_t = ttest_rel(m, b).pvalue
                try:
                    p_w = wilcoxon(m, b).pvalue
                except:
                    p_w = np.nan

                row[f"{metric}_mean"] = mean_m
                row[f"{metric}_std"] = std_m
                row[f"{metric}_ci_low"] = ci_low
                row[f"{metric}_ci_high"] = ci_high
                row[f"{metric}_p_ttest"] = p_t
                row[f"{metric}_p_wilcoxon"] = p_w

            df_all.append(row)

        out_path = os.path.join(output_dir, f"task5_statistical_summary_{dataset_base}.csv")
        pd.DataFrame(df_all).to_csv(out_path, index=False)



BASELINE = "lightautoml"
METRICS = ["accuracy", "f1", "precision", "recall"]

build_task5_statistics_A(CV_FOLDS_METRICS_PATH, TASK5_COMPARISONS_PATH, BASELINE, METRICS)
```
