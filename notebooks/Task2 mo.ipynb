```python
def concat_summaries(metrics_path, new_summary_path):
    summaries_dict, _ = loop_data_files(metrics_path)
    data_names = [i.split('.')[0] for i in data_dict_classification_only.keys()]
    dfs_names = list(summaries_dict.keys())

    for data_name in data_names:
        summary_rows = []
        matching_files = [f for f in dfs_names if data_name in f]
        for seed in RANDOM_SEEDS:
            for fname in matching_files:
                df = summaries_dict[fname]
                rows = df[df['seed'] == seed]
                for a, r in rows.iterrows():
                    row_dict = {
                        'dataset': r.get('dataset', data_name),
                        'seed': r.get('seed', seed),
                        'framework': fname.split('_')[0],
                        'accuracy': round(r.get('accuracy'), 4),
                        'f1': round(r.get('f1'), 4),
                        'precision': round(r.get('precision'), 4),
                        'recall': round(r.get('recall'), 4),
                        'training_runtime_sec': r.get('training_runtime_sec'),
                        'inference_runtime_sec': r.get('inference_runtime_sec'),
                        'best_params': r.get('best_params')
                    }
                    summary_rows.append(row_dict)
        comparison_summary = pd.DataFrame(summary_rows)
        out_path = os.path.join(new_summary_path, f"task2_boostings_tuned_comparison_{data_name}.csv")
        comparison_summary.to_csv(out_path, index=False)

concat_summaries(BOOSTINGS_TUNED_PATH, TASK2_COMPARISONS_PATH)
```
