```python
def concat_summaries(metrics_path, new_summary_path):
    summaries_dict, eda_dict = loop_data_files(metrics_path)
    data_names = [i.split('.')[0] for i in data_dict_classification_only.keys()]
    dfs_names = list(summaries_dict.keys())

    for data_name in data_names:
        summary_rows = []
        matching_files = [file_name for file_name in dfs_names if data_name in file_name]
        print(matching_files)

        for seed in RANDOM_SEEDS:
            for fname in matching_files:
                df = summaries_dict[fname]
                if 'seed' not in df.columns:
                    continue  

                rows = df[df['seed'] == seed]

                for a, r in rows.iterrows():
                    row_dict = {
                        'dataset': r.get('dataset', data_name),
                        'seed': r.get('seed', seed),
                        'framework': fname.split('_')[0],
                        'accuracy': round(r.get('accuracy', None), 4) if r.get('accuracy', None) is not None else None,
                        'f1': round(r.get('f1', None), 4) if r.get('f1', None) is not None else None,
                        'precision': round(r.get('precision', None), 4) if r.get('precision', None) is not None else None,
                        'recall': round(r.get('recall', None), 4) if r.get('recall', None) is not None else None,
                        'runtime': round(r.get('runtime', None), 4) if r.get('runtime', None) is not None else None
                    }

                    if 'fold' in df.columns:
                        row_dict['fold'] = r['fold']

                    summary_rows.append(row_dict)

        comparison_summary = pd.DataFrame(summary_rows)

        out_path = os.path.join(new_summary_path, f"task1_automl_comparison_{data_name}.csv")
        comparison_summary.to_csv(out_path, index=False)

concat_summaries(AUTOML_METRICS_PATH, TASK1_COMPARISONS_PATH)
```
