```python
def concat_summaries(metrics_path, new_summary_path):
    summaries_dict, eda_dict = loop_data_files(metrics_path)
    data_names = [i for i in data_dict_classification_only.keys()]
    dfs_names = list(summaries_dict.keys())
    dfs_values = list(summaries_dict.values())

    for data in data_names:

        summary_rows = []
        dataset_base = data.replace(".csv", "")
        matching_files = [file_name for file_name in dfs_names if dataset_base in file_name]
        for seed in RANDOM_SEEDS:
            for file_name in matching_files:
                df = summaries_dict[file_name]
                df['framework'] = file_name.split('_')[0]
                df_seed = df[df['seed'] == seed]

                for framework in df_seed['framework'].unique():
                    df_f = df_seed[df_seed['framework'] == framework]

                    mean_acc, std_acc = df_f['accuracy'].mean(), df_f['accuracy'].std()
                    mean_f1, std_f1 = df_f['f1'].mean(), df_f['f1'].std()
                    mean_prec, std_prec = df_f['precision'].mean(), df_f['precision'].std()
                    mean_rec, std_rec = df_f['recall'].mean(), df_f['recall'].std()
                    mean_runtime = df_f['runtime'].mean()

                    summary_rows.append({
                        'dataset': data,   
                        'seed': seed,
                        'framework': framework,
                        'accuracy_SD_': f"{mean_acc:.5f} ± {std_acc:.5f}",
                        'f1_SD_': f"{mean_f1:.5f} ± {std_f1:.5f}",
                        'precision_SD_': f"{mean_prec:.5f} ± {std_prec:.5f}",
                        'recall_SD_': f"{mean_rec:.5f} ± {std_rec:.5f}",
                        'avg_runtime_s_': round(mean_runtime, 5)
                    })

        comparison_summary = pd.DataFrame(summary_rows)  
        out_path = os.path.join(new_summary_path, f"task3_cv_folds_comparison_{dataset_base}.csv")
        comparison_summary.to_csv(out_path, index=False)


concat_summaries(CV_FOLDS_METRICS_PATH, TASK3_COMPARISONS_PATH)
```
